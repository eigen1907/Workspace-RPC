```python
import numpy as np
import mplhep as mh
import matplotlib.pyplot as plt

mh.style.use(mh.styles.CMS)
fig, ax = plt.subplots(figsize=(16, 10))

data1 = np.random.randn(10000)
data2 = 4.0*np.random.randn(10000) - 2.0

count1, bins1 = np.histogram(data1)
count2, bins2 = np.histogram(data2)


mh.histplot(
    count1,
    bins = bins1,
    ax = ax,
    yerr = False,
    histtype = "fill",
    facecolor = 'blue',
    edgecolor = 'black',
    linewidth = 1.6,
    flow = None,
    label = f"2022",
    alpha = 0.5,
    hatch = "///",
    density = True,
)


mh.histplot(
    count2,
    bins = bins2,
    ax = ax,
    yerr = False,
    histtype = "fill",
    facecolor = 'red',
    edgecolor = 'black',
    linewidth = 1.6,
    flow = None,
    label = f"2023",
    alpha = 0.5,
    hatch = "///",
    density = True,
)


legend = ax.legend(loc=(0.65, 0.85), #xycoords='axes fraction', 
                   fontsize=28, alignment='right')

ax.annotate(f"1.952", xy=(0.95, 0.85),
            xycoords='axes fraction', 
            fontsize=28, horizontalalignment='right')
```
